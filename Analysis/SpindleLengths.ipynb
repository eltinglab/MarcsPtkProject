```python
import math
import matplotlib.pyplot as plt
import numpy as np
import os
import scipy
import seaborn as sb
```

```python
path = r'E:\KFiberSplaying\OtherStuff\SpindleLengths\TxtDocs\Ablations\Final'
lengths = []
    
for files in os.scandir(path):
        
    data = open(files, 'r')
    fileLengths = []
        
    for line in data.readlines():
            
        iData = line.split()
        valueWorks = True
            
        try:
            float(iData[0])
        except:
            valueWorks = False
            
        if valueWorks == True:
            fileLengths.append(float(iData[0]))
        
    lengths.append(fileLengths)



avgsTemp = []
stdDevsTemp = []
stdErrorsTemp = []

for i in range(len(lengths)):
    avgsTemp.append(np.mean(lengths[i]))
    stdDevsTemp.append(np.std(lengths[i]))
    stdErrorsTemp.append(np.std(lengths[i])/np.sqrt(len(lengths[i])))

avgs = [avgsTemp[0], avgsTemp[3], avgsTemp[4], avgsTemp[5], avgsTemp[2], avgsTemp[1]]
stdDevs = [stdDevsTemp[0], stdDevsTemp[3], stdDevsTemp[4], stdDevsTemp[5], stdDevsTemp[2], stdDevsTemp[1]]
stdErrors = [stdErrorsTemp[0], stdErrorsTemp[3], stdErrorsTemp[4], stdErrorsTemp[5], stdErrorsTemp[2], stdErrorsTemp[1]]

avgsK = [avgsTemp[0], avgsTemp[5], avgsTemp[2], avgsTemp[1]]
stdDevsK = [stdDevsTemp[0], stdDevsTemp[5], stdDevsTemp[2], stdDevsTemp[1]]
stdErrorsK = [stdErrorsTemp[0], stdErrorsTemp[5], stdErrorsTemp[2], stdErrorsTemp[1]]

avgsKif15 = [avgsTemp[0], avgsTemp[2], avgsTemp[1]]
stdDevsKif15 = [stdDevsTemp[0], stdDevsTemp[2], stdDevsTemp[1]]
stdErrorsKif15 = [stdErrorsTemp[0], stdErrorsTemp[2], stdErrorsTemp[1]]



lengthsAll = [lengths[0], lengths[3], lengths[4], lengths[5], lengths[2], lengths[1]]
swarmTypes = []
swarmLengths = []
barTypes = []
std = []
error = []

for i in range(len(lengthsAll)):
    
    std.append([avgs[i], avgs[i]-2*stdDevs[i], avgs[i]+2*stdDevs[i]])
    error.append([avgs[i], avgs[i]-2*stdErrors[i], avgs[i]+2*stdErrors[i]])
    
    for j in range(3):
        if i==0:
            barTypes.append('Control')
        if i==1:
            barTypes.append('Nocodazole')
        if i==2:
            barTypes.append('NuMA siRNA')
        if i==3:
            barTypes.append('STLC')
        if i==4:
            barTypes.append('Kif15-IN-1')
        if i==5:
            barTypes.append('GW108X')
    
    for j in range(len(lengthsAll[i])):
        if i==0:
            swarmTypes.append('Control')
            swarmLengths.append(lengthsAll[i][j])
        if i==1:
            swarmTypes.append('Nocodazole')
            swarmLengths.append(lengthsAll[i][j])
        if i==2:
            swarmTypes.append('NuMA siRNA')
            swarmLengths.append(lengthsAll[i][j])
        if i==3:
            swarmTypes.append('STLC')
            swarmLengths.append(lengthsAll[i][j])
        if i==4:
            swarmTypes.append('Kif15-IN-1')
            swarmLengths.append(lengthsAll[i][j])
        if i==5:
            swarmTypes.append('GW108X')
            swarmLengths.append(lengthsAll[i][j])



lengthsK = [lengths[4], lengths[3], lengths[1], lengths[0]]
swarmTypesK = []
swarmLengthsK = []
barTypesK = []
stdK = []
errorK = []

for i in range(len(lengthsK)):
    
    stdK.append([avgsK[i], avgsK[i]-2*stdDevsK[i], avgsK[i]+2*stdDevsK[i]])
    errorK.append([avgsK[i], avgsK[i]-2*stdErrorsK[i], avgsK[i]+2*stdErrorsK[i]])
    
    for j in range(3):
        if i==0:
            barTypesK.append('Control')
        if i==1:
            barTypesK.append('STLC')
        if i==2:
            barTypesK.append('Kif15-IN-1')
        if i==3:
            barTypesK.append('GW108X')
    
    for j in range(len(lengthsK[i])):
        if i==0:
            swarmTypesK.append('Control')
            swarmLengthsK.append(lengthsK[i][j])
        if i==1:
            swarmTypesK.append('STLC')
            swarmLengthsK.append(lengthsK[i][j])
        if i==2:
            swarmTypesK.append('Kif15-IN-1')
            swarmLengthsK.append(lengthsK[i][j])
        if i==3:
            swarmTypesK.append('GW108X')
            swarmLengthsK.append(lengthsK[i][j])



lengthsKif15 = [lengths[4], lengths[1], lengths[0]]
swarmTypesKif15 = []
swarmLengthsKif15 = []
barTypesKif15 = []
stdKif15 = []
errorKif15 = []

for i in range(len(lengthsKif15)):
    
    stdKif15.append([avgsKif15[i], avgsKif15[i]-2*stdDevsKif15[i], avgsKif15[i]+2*stdDevsKif15[i]])
    errorKif15.append([avgsKif15[i], avgsKif15[i]-2*stdErrorsKif15[i], avgsKif15[i]+2*stdErrorsKif15[i]])
    
    for j in range(3):
        if i==0:
            barTypesKif15.append('Control')
        if i==1:
            barTypesKif15.append('Kif15-IN-1')
        if i==2:
            barTypesKif15.append('GW108X')
    
    for j in range(len(lengthsKif15[i])):
        if i==0:
            swarmTypesKif15.append('Control')
            swarmLengthsKif15.append(lengthsKif15[i][j])
        if i==1:
            swarmTypesKif15.append('Kif15-IN-1')
            swarmLengthsKif15.append(lengthsKif15[i][j])
        if i==2:
            swarmTypesKif15.append('GW108X')
            swarmLengthsKif15.append(lengthsKif15[i][j])



NDZlenT, NDZlenP = scipy.stats.ttest_ind(lengthsAll[0], lengthsAll[1])
NuMAlenT, NuMAlenP = scipy.stats.ttest_ind(lengthsAll[0], lengthsAll[2])
STLClenT, STLClenP = scipy.stats.ttest_ind(lengthsAll[0], lengthsAll[3])
KN1lenT, KN1lenP = scipy.stats.ttest_ind(lengthsAll[0], lengthsAll[4])
K647lenT, K647lenP = scipy.stats.ttest_ind(lengthsAll[0], lengthsAll[5])

Kif15lenT, Kif15lenP = scipy.stats.ttest_ind(lengthsAll[4], lengthsAll[5])
STLC1lenT, STLC1lenP = scipy.stats.ttest_ind(lengthsAll[2], lengthsAll[3])
STLC2lenT, STLC2lenP = scipy.stats.ttest_ind(lengthsAll[2], lengthsAll[4])

lenPs = [str('%.2E'%NDZlenP), str('%.2f'%NuMAlenP), str('%.2E'%STLClenP), str('%.2E'%KN1lenP), str('%.2E'%K647lenP)]
lenPsK = [str('%.2E'%STLClenP), str('%.2E'%KN1lenP), str('%.2E'%K647lenP)]
lenPsKif15 = [str('%.2E'%KN1lenP), str('%.2E'%K647lenP)]
lenPsSTLC = [str('%.2E'%STLC1lenP), str('%.2E'%STLC2lenP)]



xs = np.array([0,1,2,3,4,5])
labels = ['Control', 'Nocodazole', 'NuMA siRNA', 'STLC', 'Kif15-IN-1', 'GW108X']
colors = ['blue', 'black', 'red', 'olive', 'limegreen', 'forestgreen']
boxColors = ['lightskyblue', 'lightgray', 'lightcoral', 'palegoldenrod', 'palegreen', 'lightgreen']

xsK = np.array([0,1,2,3])
labelsK = ['Control', 'STLC', 'Kif15-IN-1', 'GW108X']
colorsK = ['blue', 'olive', 'limegreen', 'forestgreen']
boxColorsK = ['lightskyblue', 'palegoldenrod', 'palegreen', 'lightgreen']

xsKif15 = np.array([0,1,2])
labelsKif15 = ['Control', 'Kif15-IN-1', 'GW108X']
colorsKif15 = ['blue', 'limegreen', 'forestgreen']
boxColorsKif15 = ['lightskyblue', 'palegreen', 'darkseagreen']

print(avgs)
```
